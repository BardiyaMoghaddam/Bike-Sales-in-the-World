```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns




file_path ='/Users/bardiyamoghaddam/Library/CloudStorage/OneDrive-USN/private stuff/Data analyst/Excel/Excel portfolio to be sent to employers/Bike Sales Dataset.xlsx'
df = pd.read_excel(file_path)

#First I wanna get an overview of my dataset

df.head()
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
#1- their education level

education_counts = df['Education'].value_counts()
print("Education Level of Most Bike Buyers:\n", education_counts)


# Visualizing 
sns.countplot(data=df, x='Education', order=df['Education'].value_counts().index)
plt.title('Education Level of Bike Buyers')
plt.xticks(rotation=45)
```

```python
region_counts = df['Region'].value_counts()
print("\nRegions of Most Bike Buyers:\n", region_counts)

# Visualizing 
sns.countplot(data=df, x='Region', order=df['Region'].value_counts().index)
plt.title('Regions of Bike Buyers')
plt.xticks(rotation=45)
```

```python
income_counts = df['Income'].value_counts().head(5)
print("\nIncome Level of Most Bike Buyers:\n", income_counts)

plt.figure(figsize=(10, 6))
income_counts.plot(kind='bar', color='darkred', edgecolor='black')
plt.title('Top 5 Most Common Income Levels of Bike Buyers')
plt.xlabel('Income Level')
plt.ylabel('Number of Buyers')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
gender_counts = df['Gender'].value_counts()
print("\nGender Comparison in Bike Buying:\n", gender_counts)

df['Gender'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Gender Distribution of Bike Buyers')
plt.ylabel('')
```

```python
occupation_counts = df['Occupation'].value_counts()
print("\nOccupations of Bike Buyers:\n", occupation_counts)

sns.countplot(data=df, x='Occupation', order=df['Occupation'].value_counts().index)
plt.title('Occupations of Bike Buyers')
plt.xticks(rotation=45)
```

```python
age_distribution = df['Age'].describe()
print("\nAge Distribution of Bike Buyers:\n", age_distribution)


plt.figure(figsize=(10, 6))
sns.kdeplot(df['Age'], fill=True)
plt.title('Age Distribution of Bike Buyers')
plt.xlabel('Age')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

```python
Status_counts = df['Marital Status'].value_counts()
print("\nMarital Status of Bike Buyers:\n", Status_counts)

sns.countplot(data=df, x='Marital Status', order=df['Marital Status'].value_counts().index)
plt.title('Marital Status of Bike Buyers')
plt.xticks(rotation=45)
```

```python
#Marriage vs. purchasing bike

df['Purchased_Bike_df'] = df['Purchased Bike'].map({'Yes': 1, 'No': 0})


Marital_impact = df.groupby('Marital Status')['Purchased_Bike_df'].mean().reset_index()

# Visualizing the result
plt.figure(figsize=(10, 6))
sns.barplot(data=Marital_impact, x='Marital Status', y='Purchased_Bike_df')
plt.title('Marital Status vs. Bike Purchases')
plt.xlabel('Marital Status')
plt.ylabel('Proportion of Bike Purchases')
plt.show()
```

```python
# Car Ownership vs. Bike Purchases

df['Purchased_Bike_df'] = df['Purchased Bike'].map({'Yes': 1, 'No': 0})


car_ownership_impact = df.groupby('Cars')['Purchased_Bike_df'].mean().reset_index()

# Visualizing the result
plt.figure(figsize=(10, 6))
sns.barplot(data=car_ownership_impact, x='Cars', y='Purchased_Bike_df')
plt.title('Car Ownership vs. Bike Purchases')
plt.xlabel('Number of Cars Owned')
plt.ylabel('Proportion of Bike Purchases')
plt.show()
```

```python
# Education level vs. Bike Purchases

df['Purchased_Bike_df'] = df['Purchased Bike'].map({'Yes': 1, 'No': 0})

Education_impact = df.groupby('Education')['Purchased_Bike_df'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(data=Education_impact, x='Education', y='Purchased_Bike_df')
plt.title('Education Level vs. Bike Purchases')
plt.xlabel('Education Level')
plt.ylabel('Proportion of Bike Purchases')
plt.show()

```

```python
###age vs. Bike Purchases?

purchased_bike_df = df[df['Purchased Bike'] == 'Yes']

# Check if the 'Age' column exists and has valid data
if 'Age' in purchased_bike_df.columns and not purchased_bike_df['Age'].isnull().all():
    # Create a histogram the age distribution
    plt.figure(figsize=(10, 6))
    sns.histplot(purchased_bike_df['Age'], kde=True, bins=20, edgecolor='black')
    plt.title('Age Distribution of Bike Buyers')
    plt.xlabel('Age')
    plt.ylabel('Number of Purchases')
    plt.show()
```

```python
###region vs. purchasing bike?

region_impact = purchased_bike_df['Region'].value_counts()

# Create a pie chart
plt.figure(figsize=(10, 6))
region_impact.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Bike Purchases by Region')
plt.ylabel('')  # Hide the y-label for cleaner appearance
plt.show()
```
